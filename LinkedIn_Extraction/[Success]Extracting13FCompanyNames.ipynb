# Extracting the Names for Processing

```python
import os
from dotenv import load_dotenv
from pymongo import MongoClient

# Load environment variables
load_dotenv()

# Get MongoDB URI from .env file
mongodb_uri = os.getenv('13F_MongoDB_URI')

# Connect to MongoDB
client = MongoClient(mongodb_uri)

# Access the database and collection
db = client['13f_filings']
collection = db['investment_firms']

# Extract firm names from the collection
firm_names = []
for document in collection.find({}, {'Firm Name': 1, '_id': 0}):
    if 'Firm Name' in document:
        firm_names.append(document['Firm Name'])

# Print the extracted firm names
print("Extracted firm names:")
for name in firm_names:
    print(name)

# Close the MongoDB connection
client.close()
```
